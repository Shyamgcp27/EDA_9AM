**ENCODING**
- It is tye process of converting caregorical data to numerical data
- ML/DL or any models works based on mathematics
- Models cannot understand English characters
- So it is very important to convert categorical label to numerical values
- For example, gender column will have two labels 
  - Male
  - Female
- Here Female will be represented with 0 and Male will be represented with 1 based on Alphabetical order.

```python
visa_df['case_status'].value_counts()
```

```python
visa_df['case_status'].unique()
```

- Here...
- Certified should assign as 0
- Denied should assign as 1

```python
# Mapping
# Here we use dictionary method as it has key value pair
d={'Certified':0,'Denied':1}
d
```

```python
# Lets take the total case_status data
visa_df['case_status']
```

```python
# Now lete apply 0 and 1 to the case_status data
visa_df['case_status'].map(d)
```

```python
# Updating the visa_df case_status 
visa_df['case_status']=visa_df['case_status'].map(d)
visa_df
```

# Label-Encoder

- It is a method that converts calegorical labels to numerical values
- It is under package: **sklearn(sickit-learn)**
  - The class name: **preprocessing**
      - The method name: **LabelEncoder**

```python
# from <package>.<class> import <method>

from sklearn.preprocessing import LabelEncoder
```

**Steps to Follow while using the Label-Encoder**
- step-1: import the methos
- step-2: save the method
- step-3:Apply fit transform

- **fit-transform**
  - **fit**: means finding the logic
    - in the above example fit means create **mapper** on **dictionary**
  - **transform**: apply the logic and change the data
    - in the above example transform means **applying the mapper on the data**

```python
# Applying it for case_status column
visa_df=pd.read_csv(r"C:\Users\Amma\Documents\NareshIT\DataFiles\Visadataset.csv")
cat=visa_df.select_dtypes(include='object').columns
# step-2
from sklearn.preprocessing import LabelEncoder

# step-3: Save the method
le=LabelEncoder()

# Apply fit-transform
le.fit_transform(visa_df['case_status'])
visa_df['case_status']=le.fit_transform(visa_df['case_status'])
visa_df
```

```python
# Applying it for continent column

visa_df=pd.read_csv(r"C:\Users\Amma\Documents\NareshIT\DataFiles\Visadataset.csv")
cat=visa_df.select_dtypes(include='object').columns
# step-2
from sklearn.preprocessing import LabelEncoder

# step-3: Save the method
le=LabelEncoder()

# step-4: Apply fit-transform
le.fit_transform(visa_df['continent'])
visa_df['continent']=le.fit_transform(visa_df['continent'])
visa_df
```

```python
# Applying on all categorical columns

visa_df=pd.read_csv(r"C:\Users\Amma\Documents\NareshIT\DataFiles\Visadataset.csv")
cat=visa_df.select_dtypes(include='object').columns
# step-2
from sklearn.preprocessing import LabelEncoder

# step-3: Save the method
le=LabelEncoder()

# step-4: Apply fit-transform
for i in cat[1:]:
    le.fit_transform(visa_df[i])
    visa_df[i]=le.fit_transform(visa_df[i])
visa_df
```

**OneHotEncoder**

- In this method, First how many labels are the in the column that mnay new columns will be created
- For example: case_status has two labels
  - Certified
  - Denied
- So it will create two new extra columns as
  - case_status_Certified
  - case_status_Denied
- So, at a time one column value become 1 and other column value becomes 0
- This is **one hot encoder**
- Example:
|case_status|case_status_Certified|case_status_Denied|
|-|-|-|
|Certified|1|0|
|Denied|0|1|
|Denied|0|1|
|Certified|1|0|
