```python
wage_data=visa_df['prevailing_wage']
q1=np.percentile(pwage,25)
q2=np.percentile(pwage,50)
q3=np.percentile(pwage,75)
iqr=q3-q1
lb=q1-1.5*iqr
ub=q3+1.5*iqr
lb,ub
con1=wage_data<lb
con2=wage_data>ub
con=con1 | con2   # the line between con1, con2 indicates 'or'
visa_df[con]
len(visa_df[con])
```

```python
outliers_df=visa_df[con]
outliers_df['prevailing_wage']
outliers_df['prevailing_wage'].values
outliers_list=outliers_df['prevailing_wage'].values.tolist()
outliers_list
```

```python
# call each observation of wage data
# if that observation is available in these 427 values then that observation will be replaced by median
# otherwise that observation will be shown as it is

wage_median=visa_df['prevailing_wage'].median()
wage_data=visa_df['prevailing_wage']
new_wage_data=[]
for i in wage_data:
    if i in outliers_list:
        new_wage_data.append(wage_median)
    else:
        new_wage_data.append(i)
new_wage_data
```

**winsorization**

- whatever the outliers less than lb, fill those with lb
- whatever the outliers greater than ub, fill those with ub
- for this we have a method in numpy called **clip**

```python
# step-1: get wage data values
# step-2:calculate lb
# step-3:calculate ub
# step-4:check the parameters and apply in nump as np.clip()
# step-5:again implement the above visualization analysis
```

```python
wage_data=visa_df['prevailing_wage']
q1=np.percentile(pwage,25)
q2=np.percentile(pwage,50)
q3=np.percentile(pwage,75)
iqr=q3-q1
lb=q1-1.5*iqr
ub=q3+1.5*iqr
wage_data_clip=np.clip(wage_data,lb,ub)
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2,2,1).hist(wage_data)
plt.subplot(2,2,2).hist(wage_data_clip)
plt.subplot(2,2,3).boxplot(wage_data,vert=False)
plt.subplot(2,2,4).boxplot(wage_data_clip,vert=False)
plt.show()
```
