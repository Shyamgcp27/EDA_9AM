**Step-13: Dropping the column**

- labels
- axis
- index
- columns
- inplace

```python
# If we want to drop an index(row) then we need to use index argument name
df5.drop(index=1)
```

```python
# But after the above step if we check the original df
# We see all data
df5
```

```python
# If we want to drop a column then we need to use column name
df5.drop(columns='marks')
```

```python
# But again after the above step if we check the original df
# We see all data
df5
```

```python
# To drop multiple columns we need to give the column names in a list
df5.drop(columns=['marks','City'])
```

```python
# But again after the above step if we check the original df
# We see all data
df5
```

```python
# Python won't understand the below 2 cases of no arguments
df5.drop('marks') # if argument is not given it will given label
df5.drop(1) # here 1 become label as argument is not given

# so to make python understand whether the given value/dats is whether a column or index, we need to use 'Axis'
# By default --> Axis=0 (represents index)
# Means if we dont mention any Axis values by default it will be taken as zero(0)
# Axis=0 indicates index(row)
# Axis=1 indicates columns
```

```python
# this will error because there is values as 'marks' in index/rows
df5.drop('marks')
```

```python
df5.drop('marks',axis=0) # still gives error as axis=0 indicates index
```

```python
# So the correct code to drop the column is
df5.drop('marks',axis=1)
```

```python
df5
```

**IN-PLACE**

- When we perform any operation the result/output is not overwritting in the above cases
- Means the original data is still existing in the dataframe even after dropping the index or columns
- In oreder to correct this scenarion we need to use **inplace** argument
- By default **inplace=Fault**
- Make **inplace=True** if we want to delete data permanantly or overwrite the result

```python
# As we are taking only one column, we can give it directly
# If multiple columbs has to be dropped then mention them in a list
df5.drop('marks',axis=1,inplace=True)
df5
```

```python
# Now we can see the marks column got deleted permanently
df5
```
